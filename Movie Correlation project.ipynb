```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\Acer\Desktop\Data Analyst project\Python project\movies.csv')
```

```python
#Look at the data

df.head()
```

```python
#checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#data types for columns

df.dtypes
```

```python
#changing data type of columns

df['gross'] = pd.to_numeric(df['gross'], errors='coerce', downcast='integer')
df['gross'].isna().sum()
df[df['gross'].isna()]
df['gross'].fillna(0, inplace=True)
df['budget'] = df['budget'].fillna(0)
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].round().astype('int64')
```

```python
df
```

```python
#create correct year column 

df['yearcorrect'] = df['year'].astype(str).str[:4]

df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Drop duplicates

df['company'].drop_duplicates().sort_values(ascending= False)
df.head()
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix= df_numerized.corr(numeric_only=True)

plt.title('Correlation matrix for numeric features')

plt.ylabel('Movie features')

plt.xlabel("Movie features")

sns.heatmap(correlation_matrix, annot=True)

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]

high_correlation
```
